## Simulation of  modified Stuart-Landau (MSL) model with (generalized) phase model

### Model definition 

```julia
dxdt(X, I, λ=5.0) = λ*(X[1]-I) - exp(I)*X[2]-λ*((X[1]-I)^2+X[2]^2)*(X[1]-I)
dydt(X, I, λ=5.0) = exp(I)*(X[1]-I) + λ*X[2]-λ*((X[1]-I)^2+X[2]^2)*X[2]
F(X, I) = [dxdt(X, I), dydt(X, I)]

D, N = 2, 2 # number of dimensions, units
Nθ = 1000
G(X, K) = K * ((1 .- eye(N)) * X[:, 1])
dt = 1e-3; T = 100.0;
```

```julia
@time Ts, ω, Xs = find_stable_periodic_solution(F, 0, D, Nθ, [1, 0], dt, Tsit5(), 2, 0.0, print_progress=false)
println("Ts=", Ts, " (sec) , ω=", ω, " (Hz)")
```

```julia
# F, Imin, Imax, dI, D, Nθ, nothing, dt, alg, origin_val_idx, origin_thr
ωI, ζθI, ξθI, XsI = generalized_phase_sensitivity_func(F, -1.2, 1.2, 0.1, D, Nθ, nothing, dt, Tsit5(), 2, 0.0);
```

```julia
function coupled_func!(dX, X, p, t)
    g, κ = p
    Iext = g(X) # input
    for i in 1:size(dX)[1]
        dX[i, :] = κ[i] * F(X[i, :], Iext[i])
    end
end
```

```julia
function all_simulation(N, D, T, ωI, ζθI, ξθI, XsI, coupled_func!, G, K, κ, initθ, alg=Vern9(), NΘ=100)
    g(X) = G(X, K)
    IΘ = compute_IsΘ(g, N, D, XsI, NΘ)
    
    _, Xos, Θgos, Θcos = coupled_original_system(N, D, T, XsI, g, coupled_func!, initθ, κ, alg)
    _, Xcpm, Θcpm = coupled_conventinal_phase_model(N, D, T, XsI, g, ωI, ζθI, Θcos[1, :], κ, alg)
    _, Xgpm, Θgpm = coupled_generalized_phase_model_I(N, D, T, XsI, IΘ, ωI, ξθI, Θgos[1, :], κ, alg)
    return Xos, Xcpm, Xgpm, Θgos, Θcos, Θcpm, Θgpm
end
```

```julia
K = 0.3
κs = [[1.0, 1.001], [1.0, 1.01], [1.0, 1.1]]
initθ = [0, 0];
Xoss, Xcpms, Xgpms = Any[], Any[], Any[]

@time for κ in κs
    res = all_simulation(N, D, T, ωI, ζθI, ξθI, XsI, coupled_func!, G, K, κ, initθ, alg)
    push!(Xoss, res[1])
    push!(Xcpms, res[2])
    push!(Xgpms, res[3])
end
```
